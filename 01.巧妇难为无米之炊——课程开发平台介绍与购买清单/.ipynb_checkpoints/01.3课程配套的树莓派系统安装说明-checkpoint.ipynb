# 课程配套的树莓派系统安装说明

课程配套树莓派系统安装只需三步。

1.下载课程配套的树莓派系统镜像和windows版本的TF卡烧写工具

百度云盘地址为:

    链接：https://pan.baidu.com/s/14KYZl21Hyf3sjjcWxa5_bg 
    提取码：dbe5

包含系统镜像文件和烧写工具。

烧写工具名字为:win32diskimager-1.0.0-install.exe


2.安装烧写工具软件，进行镜像烧写

在windows下点击“win32diskimager-1.0.0-install.exe”，安装“win32diskimager”软件。

安装完毕后，运行“win32diskimager”软件，选择我们要烧写的镜像文件进行烧写。

3.启动系统，配置系统

将TF卡插入树莓派4B上，连接电源线、显示屏视频线、麦克风线、音箱线、鼠标和键盘线，启动系统。

使用raspi-config命令让系统使用所有TF卡容量。

将树莓派连接到无线网中。

运行桌面上的“开发板外设测试程序”，进行课程开发环境的监测。
